```python
chunks = pd.read_csv('bookings.csv.bz2', "^", chunksize=1000)
resultados = []
for chunk in chunks:
    resultados.append(chunk[['arr_port', 'pax']][chunk['year'] == 2013].groupby(['arr_port'], as_index=False).sum())
    
```

```python
arrejuntados = pd.concat(resultados)
arrejuntados
```

```python
final = arrejuntados.groupby(['arr_port'], as_index=False).sum().sort_values(by='pax', ascending=False).head(10)
final
```
