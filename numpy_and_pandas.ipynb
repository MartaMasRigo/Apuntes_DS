```python
X = np.arange(1, 10, .05).reshape(-1, 1)
y = np.array([x + np.random.normal(0, x / 5) for x in X]).reshape(-1, 1)
```

```python
X.shape
```

```python
df= pd.DataFrame({"X":X.squeeze(), "y":y.squeeze()})
```

```python
df.plot(kind='scatter', x='X', y='y')
```

```python
# Calcular la pendiente:

num=((df['X'] - df['X'].mean()) * (df['y'] - df['y'].mean())).sum()
den=((df['X'] - df['X'].mean())**2).sum()
slope = num/den
print('La pendiente es:{:.2f}'.format(slope))
```

```python
# Calcular el sesgo:
bias= df['X'].mean() - slope * df['X'].mean()
print('El sesgo es de: {:.2f}'.format(bias))
```

```python
ax= df.plot(kind='scatter', x='X', y='y')
ax.plot([df['X'].min(), df['X'].max()], [slope*df['X'].min()+bias, slope*df['X'].max()+bias], lw=4, color='r')
```
